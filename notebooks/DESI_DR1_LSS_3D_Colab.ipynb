## Load a research bundle

Paste a GitHub Release asset URL into `BUNDLE_URL` after publication, or leave it blank and upload the Parquet file directly.

```python
from pathlib import Path
import requests

BUNDLE_URL = ""  # GitHub Release asset URL, or leave empty for upload
BUNDLE_PATH = Path("desi_dr1_lss_research_bundle.parquet")

if BUNDLE_URL:
    response = requests.get(BUNDLE_URL, stream=True, timeout=120)
    response.raise_for_status()
    with BUNDLE_PATH.open("wb") as handle:
        for block in response.iter_content(chunk_size=1024 * 1024):
            if block:
                handle.write(block)
else:
    from google.colab import files
    uploaded = files.upload()
    BUNDLE_PATH = Path(next(iter(uploaded)))

print(f"Using {BUNDLE_PATH} ({BUNDLE_PATH.stat().st_size / 1024**2:.2f} MiB)")
```

```python
!python plot_desi_research_figure.py --input {BUNDLE_PATH} --output-dir figures --render-rows 120000

from IPython.display import Image, display
display(Image(filename="figures/desi_dr1_lss_3d_research_view.png"))
display(Image(filename="figures/desi_dr1_lss_redshift_summary.png"))
```
